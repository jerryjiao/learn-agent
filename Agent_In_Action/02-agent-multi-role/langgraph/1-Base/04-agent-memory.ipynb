```python
# üîç ÁéØÂ¢É‰ø°ÊÅØÊ£ÄÊü•ËÑöÊú¨
#
# Êú¨ËÑöÊú¨ÁöÑ‰ΩúÁî®Ôºö
# 1. ÂÆâË£Ö pandas Â∫ìÁî®‰∫éÊï∞ÊçÆË°®Ê†ºÂ±ïÁ§∫
# 2. Ê£ÄÊü•Á≥ªÁªüÁöÑÂêÑÈ°πÈÖçÁΩÆ‰ø°ÊÅØ
# 3. ÁîüÊàêËØ¶ÁªÜÁöÑÁéØÂ¢ÉÊä•ÂëäË°®Ê†º
#
# ÂØπ‰∫éÂàùÂ≠¶ËÄÖÊù•ËØ¥ÔºåËøô‰∏™Ê≠•È™§Â∏ÆÂä©‰Ω†Ôºö
# - ‰∫ÜËß£ÂΩìÂâçËøêË°åÁéØÂ¢ÉÁöÑÁ°¨‰ª∂ÈÖçÁΩÆ
# - Á°ÆËÆ§ÊòØÂê¶Êª°Ë∂≥Ê®°ÂûãËøêË°åÁöÑÊúÄ‰ΩéË¶ÅÊ±Ç
# - Â≠¶‰π†Â¶Ç‰ΩïÈÄöËøá‰ª£Á†ÅËé∑ÂèñÁ≥ªÁªü‰ø°ÊÅØ

# ÂÆâË£Ö pandas Â∫ì - Áî®‰∫éÂàõÂª∫ÂíåÂ±ïÁ§∫Êï∞ÊçÆË°®Ê†º
# pandas ÊòØ Python ‰∏≠ÊúÄÊµÅË°åÁöÑÊï∞ÊçÆÂ§ÑÁêÜÂíåÂàÜÊûêÂ∫ì
%pip install pandas==2.2.2 tabulate==0.9.0

import platform # ÂØºÂÖ• platform Ê®°Âùó‰ª•Ëé∑ÂèñÁ≥ªÁªü‰ø°ÊÅØ
import os # ÂØºÂÖ• os Ê®°Âùó‰ª•‰∏éÊìç‰ΩúÁ≥ªÁªü‰∫§‰∫í
import subprocess # ÂØºÂÖ• subprocess Ê®°Âùó‰ª•ËøêË°åÂ§ñÈÉ®ÂëΩ‰ª§
import pandas as pd # ÂØºÂÖ• pandas Ê®°ÂùóÔºåÈÄöÂ∏∏Áî®‰∫éÊï∞ÊçÆÂ§ÑÁêÜÔºåËøôÈáåÁî®‰∫éÂàõÂª∫Ë°®Ê†º
import shutil # ÂØºÂÖ• shutil Ê®°Âùó‰ª•Ëé∑ÂèñÁ£ÅÁõòÁ©∫Èó¥‰ø°ÊÅØ

# Ëé∑Âèñ CPU ‰ø°ÊÅØÁöÑÂáΩÊï∞ÔºåÂåÖÊã¨Ê†∏ÂøÉÊï∞Èáè
def get_cpu_info():
    cpu_info = "" # ÂàùÂßãÂåñ CPU ‰ø°ÊÅØÂ≠óÁ¨¶‰∏≤
    physical_cores = "N/A"
    logical_cores = "N/A"

    if platform.system() == "Windows": # Â¶ÇÊûúÊòØ Windows Á≥ªÁªü
        cpu_info = platform.processor() # ‰ΩøÁî® platform.processor() Ëé∑Âèñ CPU ‰ø°ÊÅØ
        try:
            # Ëé∑Âèñ Windows ‰∏äÁöÑÊ†∏ÂøÉÊï∞Èáè (ÈúÄË¶Å WMI)
            import wmi
            c = wmi.WMI()
            for proc in c.Win32_Processor():
                physical_cores = proc.NumberOfCores
                logical_cores = proc.NumberOfLogicalProcessors
        except:
            pass # Â¶ÇÊûú WMI ‰∏çÂèØÁî®ÔºåÂøΩÁï•ÈîôËØØ

    elif platform.system() == "Darwin": # Â¶ÇÊûúÊòØ macOS Á≥ªÁªü
        # Âú® macOS ‰∏ä‰ΩøÁî® sysctl ÂëΩ‰ª§Ëé∑Âèñ CPU ‰ø°ÊÅØÂíåÊ†∏ÂøÉÊï∞Èáè
        os.environ['PATH'] = os.environ['PATH'] + os.pathsep + '/usr/sbin' # Êõ¥Êñ∞ PATH ÁéØÂ¢ÉÂèòÈáè
        try:
            process_brand = subprocess.Popen(['sysctl', "machdep.cpu.brand_string"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_brand, stderr_brand = process_brand.communicate()
            cpu_info = stdout_brand.decode().split(': ')[1].strip() if stdout_brand else "Could not retrieve CPU info"

            process_physical = subprocess.Popen(['sysctl', "hw.physicalcpu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_physical, stderr_physical = process_physical.communicate()
            physical_cores = stdout_physical.decode().split(': ')[1].strip() if stdout_physical else "N/A"

            process_logical = subprocess.Popen(['sysctl', "hw.logicalcpu"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            stdout_logical, stderr_logical = process_logical.communicate()
            logical_cores = stdout_logical.decode().split(': ')[1].strip() if stdout_logical else "N/A"

        except:
            cpu_info = "Could not retrieve CPU info"
            physical_cores = "N/A"
            logical_cores = "N/A"

    else:  # Linux Á≥ªÁªü
        try:
            # Âú® Linux ‰∏äËØªÂèñ /proc/cpuinfo Êñá‰ª∂Ëé∑Âèñ CPU ‰ø°ÊÅØÂíåÊ†∏ÂøÉÊï∞Èáè
            with open('/proc/cpuinfo') as f:
                physical_cores_count = 0
                logical_cores_count = 0
                cpu_info_lines = []
                for line in f:
                    if line.startswith('model name'): # Êü•Êâæ‰ª• 'model name'ÂºÄÂ§¥ÁöÑË°å
                        if not cpu_info: # Âè™Ëé∑ÂèñÁ¨¨‰∏Ä‰∏™ model name
                            cpu_info = line.split(': ')[1].strip()
                    elif line.startswith('cpu cores'): # Êü•Êâæ‰ª• 'cpu cores' ÂºÄÂ§¥ÁöÑË°å
                        physical_cores_count = int(line.split(': ')[1].strip())
                    elif line.startswith('processor'): # Êü•Êâæ‰ª• 'processor' ÂºÄÂ§¥ÁöÑË°å
                        logical_cores_count += 1
                physical_cores = str(physical_cores_count) if physical_cores_count > 0 else "N/A"
                logical_cores = str(logical_cores_count) if logical_cores_count > 0 else "N/A"
                if not cpu_info:
                     cpu_info = "Could not retrieve CPU info"

        except:
            cpu_info = "Could not retrieve CPU info"
            physical_cores = "N/A"
            logical_cores = "N/A"

    return f"{cpu_info} ({physical_cores} physical cores, {logical_cores} logical cores)" # ËøîÂõû CPU ‰ø°ÊÅØÂíåÊ†∏ÂøÉÊï∞Èáè


# Ëé∑ÂèñÂÜÖÂ≠ò‰ø°ÊÅØÁöÑÂáΩÊï∞
def get_memory_info():
    mem_info = "" # ÂàùÂßãÂåñÂÜÖÂ≠ò‰ø°ÊÅØÂ≠óÁ¨¶‰∏≤
    if platform.system() == "Windows":
        # Âú® Windows ‰∏ä‰∏çÂÆπÊòìÈÄöËøáÊ†áÂáÜÂ∫ìËé∑ÂèñÔºåÈúÄË¶ÅÂ§ñÈÉ®Â∫ìÊàñ PowerShell
        mem_info = "Requires external tools on Windows" # ËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ
    elif platform.system() == "Darwin": # Â¶ÇÊûúÊòØ macOS Á≥ªÁªü
        # Âú® macOS ‰∏ä‰ΩøÁî® sysctl ÂëΩ‰ª§Ëé∑ÂèñÂÜÖÂ≠òÂ§ßÂ∞è
        process = subprocess.Popen(['sysctl', "hw.memsize"], stdout=subprocess.PIPE, stderr=subprocess.PIPE) # ËøêË°å sysctl ÂëΩ‰ª§
        stdout, stderr = process.communicate() # Ëé∑ÂèñÊ†áÂáÜËæìÂá∫ÂíåÊ†áÂáÜÈîôËØØ
        mem_bytes = int(stdout.decode().split(': ')[1].strip()) # Ëß£ÊûêËæìÂá∫ÔºåËé∑ÂèñÂÜÖÂ≠òÂ§ßÂ∞èÔºàÂ≠óËäÇÔºâ
        mem_gb = mem_bytes / (1024**3) # ËΩ¨Êç¢‰∏∫ GB
        mem_info = f"{mem_gb:.2f} GB" # Ê†ºÂºèÂåñËæìÂá∫
    else:  # Linux Á≥ªÁªü
        try:
            # Âú® Linux ‰∏äËØªÂèñ /proc/meminfo Êñá‰ª∂Ëé∑ÂèñÂÜÖÂ≠ò‰ø°ÊÅØ
            with open('/proc/meminfo') as f:
                total_mem_kb = 0
                available_mem_kb = 0
                for line in f:
                    if line.startswith('MemTotal'): # Êü•Êâæ‰ª• 'MemTotal' ÂºÄÂ§¥ÁöÑË°å
                        total_mem_kb = int(line.split(':')[1].strip().split()[0]) # Ëß£ÊûêË°åÔºåËé∑ÂèñÊÄªÂÜÖÂ≠òÔºàKBÔºâ
                    elif line.startswith('MemAvailable'): # Êü•Êâæ‰ª• 'MemAvailable' ÂºÄÂ§¥ÁöÑË°å
                         available_mem_kb = int(line.split(':')[1].strip().split()[0]) # Ëß£ÊûêË°åÔºåËé∑ÂèñÂèØÁî®ÂÜÖÂ≠òÔºàKBÔºâ

                if total_mem_kb > 0:
                    total_mem_gb = total_mem_kb / (1024**2) # ËΩ¨Êç¢‰∏∫ GB
                    mem_info = f"{total_mem_gb:.2f} GB" # Ê†ºÂºèÂåñËæìÂá∫ÊÄªÂÜÖÂ≠ò
                    if available_mem_kb > 0:
                        available_mem_gb = available_mem_kb / (1024**2)
                        mem_info += f" (Available: {available_mem_gb:.2f} GB)" # Ê∑ªÂä†ÂèØÁî®ÂÜÖÂ≠ò‰ø°ÊÅØ
                else:
                     mem_info = "Could not retrieve memory info" # Â¶ÇÊûúËØªÂèñÊñá‰ª∂Âá∫ÈîôÔºåËÆæÁΩÆÈîôËØØ‰ø°ÊÅØ

        except:
            mem_info = "Could not retrieve memory info" # Â¶ÇÊûúËØªÂèñÊñá‰ª∂Âá∫ÈîôÔºåËÆæÁΩÆÈîôËØØ‰ø°ÊÅØ
    return mem_info # ËøîÂõûÂÜÖÂ≠ò‰ø°ÊÅØ

# Ëé∑Âèñ GPU ‰ø°ÊÅØÁöÑÂáΩÊï∞ÔºåÂåÖÊã¨ÊòæÂ≠ò
def get_gpu_info():
    try:
        # Â∞ùËØï‰ΩøÁî® nvidia-smi Ëé∑Âèñ NVIDIA GPU ‰ø°ÊÅØÂíåÊòæÂ≠ò
        result = subprocess.run(['nvidia-smi', '--query-gpu=name,memory.total', '--format=csv,noheader'], capture_output=True, text=True)
        if result.returncode == 0: # Â¶ÇÊûúÂëΩ‰ª§ÊàêÂäüÊâßË°å
            gpu_lines = result.stdout.strip().split('\n') # Ëß£ÊûêËæìÂá∫ÔºåËé∑Âèñ GPU ÂêçÁß∞ÂíåÊòæÂ≠ò
            gpu_info_list = []
            for line in gpu_lines:
                name, memory = line.split(', ')
                gpu_info_list.append(f"{name} ({memory})") # Ê†ºÂºèÂåñ GPU ‰ø°ÊÅØ
            return ", ".join(gpu_info_list) if gpu_info_list else "NVIDIA GPU found, but info not listed" # ËøîÂõû GPU ‰ø°ÊÅØÊàñÊèêÁ§∫‰ø°ÊÅØ
        else:
             # Â∞ùËØï‰ΩøÁî® lshw Ëé∑ÂèñÂÖ∂‰ªñ GPU ‰ø°ÊÅØ (ÈúÄË¶ÅÂÆâË£Ö lshw)
            try:
                result_lshw = subprocess.run(['lshw', '-C', 'display'], capture_output=True, text=True)
                if result_lshw.returncode == 0: # Â¶ÇÊûúÂëΩ‰ª§ÊàêÂäüÊâßË°å
                     # ÁÆÄÂçïËß£ÊûêËæìÂá∫‰∏≠ÁöÑ product ÂêçÁß∞ÂíåÊòæÂ≠ò
                    gpu_info_lines = []
                    current_gpu = {}
                    for line in result_lshw.stdout.splitlines():
                        if 'product:' in line:
                             if current_gpu:
                                 gpu_info_lines.append(f"{current_gpu.get('product', 'GPU')} ({current_gpu.get('memory', 'N/A')})")
                             current_gpu = {'product': line.split('product:')[1].strip()}
                        elif 'size:' in line and 'memory' in line:
                             current_gpu['memory'] = line.split('size:')[1].strip()

                    if current_gpu: # Ê∑ªÂä†ÊúÄÂêé‰∏Ä‰∏™ GPU ÁöÑ‰ø°ÊÅØ
                        gpu_info_lines.append(f"{current_gpu.get('product', 'GPU')} ({current_gpu.get('memory', 'N/A')})")

                    return ", ".join(gpu_info_lines) if gpu_info_lines else "GPU found (via lshw), but info not parsed" # Â¶ÇÊûúÊâæÂà∞ GPU ‰ΩÜ‰ø°ÊÅØÊó†Ê≥ïËß£ÊûêÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ
                else:
                    return "No GPU found (checked nvidia-smi and lshw)" # Â¶ÇÊûú‰∏§‰∏™ÂëΩ‰ª§ÈÉΩÊâæ‰∏çÂà∞ GPUÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ
            except FileNotFoundError:
                 return "No GPU found (checked nvidia-smi, lshw not found)" # Â¶ÇÊûúÊâæ‰∏çÂà∞ lshw ÂëΩ‰ª§ÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ
    except FileNotFoundError:
        return "No GPU found (nvidia-smi not found)" # Â¶ÇÊûúÊâæ‰∏çÂà∞ nvidia-smi ÂëΩ‰ª§ÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ


# Ëé∑Âèñ CUDA ÁâàÊú¨ÁöÑÂáΩÊï∞
def get_cuda_version():
    try:
        # Â∞ùËØï‰ΩøÁî® nvcc --version Ëé∑Âèñ CUDA ÁâàÊú¨
        result = subprocess.run(['nvcc', '--version'], capture_output=True, text=True)
        if result.returncode == 0: # Â¶ÇÊûúÂëΩ‰ª§ÊàêÂäüÊâßË°å
            for line in result.stdout.splitlines():
                if 'release' in line: # Êü•ÊâæÂåÖÂê´ 'release' ÁöÑË°å
                    return line.split('release ')[1].split(',')[0] # Ëß£ÊûêË°åÔºåÊèêÂèñÁâàÊú¨Âè∑
        return "CUDA not found or version not parsed" # Â¶ÇÊûúÊâæ‰∏çÂà∞ CUDA ÊàñÁâàÊú¨Êó†Ê≥ïËß£ÊûêÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ
    except FileNotFoundError:
        return "CUDA not found" # Â¶ÇÊûúÊâæ‰∏çÂà∞ nvcc ÂëΩ‰ª§ÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ

# Ëé∑Âèñ Python ÁâàÊú¨ÁöÑÂáΩÊï∞
def get_python_version():
    return platform.python_version() # Ëé∑Âèñ Python ÁâàÊú¨

# Ëé∑Âèñ Conda ÁâàÊú¨ÁöÑÂáΩÊï∞
def get_conda_version():
    try:
        # Â∞ùËØï‰ΩøÁî® conda --version Ëé∑Âèñ Conda ÁâàÊú¨
        result = subprocess.run(['conda', '--version'], capture_output=True, text=True)
        if result.returncode == 0: # Â¶ÇÊûúÂëΩ‰ª§ÊàêÂäüÊâßË°å
            return result.stdout.strip() # ËøîÂõû Conda ÁâàÊú¨
        return "Conda not found or version not parsed" # Â¶ÇÊûúÊâæ‰∏çÂà∞ Conda ÊàñÁâàÊú¨Êó†Ê≥ïËß£ÊûêÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ
    except FileNotFoundError:
        return "Conda not found" # Â¶ÇÊûúÊâæ‰∏çÂà∞ conda ÂëΩ‰ª§ÔºåËÆæÁΩÆÊèêÁ§∫‰ø°ÊÅØ

# Ëé∑ÂèñÁâ©ÁêÜÁ£ÅÁõòÁ©∫Èó¥‰ø°ÊÅØÁöÑÂáΩÊï∞
def get_disk_space():
    try:
        total, used, free = shutil.disk_usage("/") # Ëé∑ÂèñÊ†πÁõÆÂΩïÁöÑÁ£ÅÁõò‰ΩøÁî®ÊÉÖÂÜµ
        total_gb = total / (1024**3) # ËΩ¨Êç¢‰∏∫ GB
        used_gb = used / (1024**3) # ËΩ¨Êç¢‰∏∫ GB
        free_gb = free / (1024**3) # ËΩ¨Êç¢‰∏∫ GB
        return f"Total: {total_gb:.2f} GB, Used: {used_gb:.2f} GB, Free: {free_gb:.2f} GB" # Ê†ºÂºèÂåñËæìÂá∫
    except Exception as e:
        return f"Could not retrieve disk info: {e}" # Â¶ÇÊûúËé∑Âèñ‰ø°ÊÅØÂá∫ÈîôÔºåËÆæÁΩÆÈîôËØØ‰ø°ÊÅØ

# Ëé∑ÂèñÁéØÂ¢É‰ø°ÊÅØ
os_name = platform.system() # Ëé∑ÂèñÊìç‰ΩúÁ≥ªÁªüÂêçÁß∞
os_version = platform.release() # Ëé∑ÂèñÊìç‰ΩúÁ≥ªÁªüÁâàÊú¨
if os_name == "Linux":
    try:
        # Âú® Linux ‰∏äÂ∞ùËØïËé∑ÂèñÂèëË°åÁâàÂíåÁâàÊú¨
        lsb_info = subprocess.run(['lsb_release', '-a'], capture_output=True, text=True)
        if lsb_info.returncode == 0: # Â¶ÇÊûúÂëΩ‰ª§ÊàêÂäüÊâßË°å
            for line in lsb_info.stdout.splitlines():
                if 'Description:' in line: # Êü•ÊâæÂåÖÂê´ 'Description:' ÁöÑË°å
                    os_version = line.split('Description:')[1].strip() # ÊèêÂèñÊèèËø∞‰ø°ÊÅØ‰Ωú‰∏∫ÁâàÊú¨
                    break # ÊâæÂà∞ÂêéÈÄÄÂá∫Âæ™ÁéØ
                elif 'Release:' in line: # Êü•ÊâæÂåÖÂê´ 'Release:' ÁöÑË°å
                     os_version = line.split('Release:')[1].strip() # ÊèêÂèñÁâàÊú¨Âè∑
                     # Â∞ùËØïËé∑Âèñ codename
                     try:
                         codename_info = subprocess.run(['lsb_release', '-c'], capture_output=True, text=True)
                         if codename_info.returncode == 0:
                             os_version += f" ({codename_info.stdout.split(':')[1].strip()})" # Â∞Ü codename Ê∑ªÂä†Âà∞ÁâàÊú¨‰ø°ÊÅØ‰∏≠
                     except:
                         pass # Â¶ÇÊûúËé∑Âèñ codename Â§±Ë¥•ÂàôÂøΩÁï•

    except FileNotFoundError:
        pass # lsb_release ÂèØËÉΩÊú™ÂÆâË£ÖÔºåÂøΩÁï•ÈîôËØØ

full_os_info = f"{os_name} {os_version}" # ÁªÑÂêàÂÆåÊï¥ÁöÑÊìç‰ΩúÁ≥ªÁªü‰ø°ÊÅØ
cpu_info = get_cpu_info() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑Âèñ CPU ‰ø°ÊÅØÂíåÊ†∏ÂøÉÊï∞Èáè
memory_info = get_memory_info() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑ÂèñÂÜÖÂ≠ò‰ø°ÊÅØ
gpu_info = get_gpu_info() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑Âèñ GPU ‰ø°ÊÅØÂíåÊòæÂ≠ò
cuda_version = get_cuda_version() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑Âèñ CUDA ÁâàÊú¨
python_version = get_python_version() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑Âèñ Python ÁâàÊú¨
conda_version = get_conda_version() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑Âèñ Conda ÁâàÊú¨
disk_info = get_disk_space() # Ë∞ÉÁî®ÂáΩÊï∞Ëé∑ÂèñÁâ©ÁêÜÁ£ÅÁõòÁ©∫Èó¥‰ø°ÊÅØ


# ÂàõÂª∫Áî®‰∫éÂ≠òÂÇ®Êï∞ÊçÆÁöÑÂ≠óÂÖ∏
env_data = {
    "È°πÁõÆ": [ # È°πÁõÆÂêçÁß∞ÂàóË°®
        "Êìç‰ΩúÁ≥ªÁªü",
        "CPU ‰ø°ÊÅØ",
        "ÂÜÖÂ≠ò‰ø°ÊÅØ",
        "GPU ‰ø°ÊÅØ",
        "CUDA ‰ø°ÊÅØ",
        "Python ÁâàÊú¨",
        "Conda ÁâàÊú¨",
        "Áâ©ÁêÜÁ£ÅÁõòÁ©∫Èó¥" # Ê∑ªÂä†Áâ©ÁêÜÁ£ÅÁõòÁ©∫Èó¥
    ],
    "‰ø°ÊÅØ": [ # ÂØπÂ∫îÁöÑ‰ø°ÊÅØÂàóË°®
        full_os_info,
        cpu_info,
        memory_info,
        gpu_info,
        cuda_version,
        python_version,
        conda_version,
        disk_info # Ê∑ªÂä†Áâ©ÁêÜÁ£ÅÁõòÁ©∫Èó¥‰ø°ÊÅØ
    ]
}

# ÂàõÂª∫‰∏Ä‰∏™ pandas DataFrame
df = pd.DataFrame(env_data)

# ÊâìÂç∞Ë°®Ê†º
print("### ÁéØÂ¢É‰ø°ÊÅØ") # ÊâìÂç∞Ê†áÈ¢ò
print(df.to_markdown(index=False)) # Â∞Ü DataFrame ËΩ¨Êç¢‰∏∫ Markdown Ê†ºÂºèÂπ∂ÊâìÂç∞Ôºå‰∏çÂåÖÂê´Á¥¢Âºï
```

```python
from langchain_openai import ChatOpenAI

def multiply(a: int, b: int) -> int:
    """Â∞Ü a ‰∏é b Áõ∏‰πò„ÄÇ

    ÂèÇÊï∞:
        a: Á¨¨‰∏Ä‰∏™Êï¥Êï∞
        b: Á¨¨‰∫å‰∏™Êï¥Êï∞
    """
    return a * b

# ËØ•ÂáΩÊï∞Â∞ÜÊ≥®ÂÜå‰∏∫‰∏Ä‰∏™ÂèØË¢´ Agent Ë∞ÉÁî®ÁöÑ‚ÄúÂ∑•ÂÖ∑‚Äù
def add(a: int, b: int) -> int:
    """Â∞Ü a ‰∏é b Áõ∏Âä†„ÄÇ

    ÂèÇÊï∞:
        a: Á¨¨‰∏Ä‰∏™Êï¥Êï∞
        b: Á¨¨‰∫å‰∏™Êï¥Êï∞
    """
    return a + b

def divide(a: int, b: int) -> float:
    """Â∞Ü a ‰∏é b Áõ∏Èô§„ÄÇ

    ÂèÇÊï∞:
        a: Á¨¨‰∏Ä‰∏™Êï¥Êï∞
        b: Á¨¨‰∫å‰∏™Êï¥Êï∞
    """
    return a / b

# Â∞ÜÂ∑•ÂÖ∑ÂàóË°®ÁªëÂÆöÂà∞Ê®°ÂûãÂÆû‰æã‰∏äÔºå‰ΩøÊ®°ÂûãËÉΩÂ§üË∞ÉÁî®ÂÆÉ‰ª¨
tools = [add, multiply, divide]
llm = ChatOpenAI(model="gpt-4o")
llm_with_tools = llm.bind_tools(tools)
```

```python
from langgraph.graph import MessagesState
from langchain_core.messages import HumanMessage, SystemMessage

# System message
sys_msg = SystemMessage(content="You are a helpful assistant tasked with performing arithmetic on a set of inputs.")

# Node
def assistant(state: MessagesState):
   return {"messages": [llm_with_tools.invoke([sys_msg] + state["messages"])]}
```

```python
from langgraph.graph import START, StateGraph
from langgraph.prebuilt import tools_condition, ToolNode
from IPython.display import Image, display

# ÊûÑÂª∫ÂõæÔºàGraphÔºâ
builder = StateGraph(MessagesState)

# ÂÆö‰πâËäÇÁÇπÔºàNodesÔºâÔºöÂÆûÈôÖÊâßË°åÂ∑•‰ΩúÁöÑÂçïÂÖÉ
builder.add_node("assistant", assistant)
builder.add_node("tools", ToolNode(tools))

# ÂÆö‰πâËæπÔºàEdgesÔºâÔºöÂÜ≥ÂÆöÊéßÂà∂ÊµÅÂ¶Ç‰ΩïÂú®ËäÇÁÇπÈó¥ÊµÅËΩ¨
builder.add_edge(START, "assistant")
builder.add_conditional_edges(
    "assistant",
    # Ëã• assistant ÊúÄÊñ∞Ê∂àÊÅØÊòØÂ∑•ÂÖ∑Ë∞ÉÁî® -> Áî± tools_condition Ë∑ØÁî±Âà∞ tools
    # Ëã•‰∏çÊòØÂ∑•ÂÖ∑Ë∞ÉÁî® -> Áî± tools_condition Ë∑ØÁî±Âà∞ END
    tools_condition,
)
builder.add_edge("tools", "assistant")
react_graph = builder.compile()

# ÊòæÁ§∫ÂõæÁªìÊûÑ
# display(Image(react_graph.get_graph(xray=True).draw_mermaid_png()))

# Â±ïÁ§∫ÂõæÁªìÊûÑ
# ÂõæÂèØËßÜÂåñ
print("ÂõæÂèØËßÜÂåñÔºö")

# ÊñπÊ°à1ÔºöÂ∞ùËØï‰ΩøÁî® Pyppeteer Êú¨Âú∞Ê∏≤ÊüìÔºàÊé®ËçêÔºâ
try:
    # ÂèØËßÜÂåñÔºöÈÄöËøá Mermaid Ê∏≤ÊüìÂõæÁªìÊûÑ
    display(Image(react_graph.get_graph().draw_mermaid_png()))
    print("‚úÖ ÂõæÊ∏≤ÊüìÊàêÂäüÔºÅ")
except Exception as e:
    print(f"‚ùå Pyppeteer Ê∏≤ÊüìÂ§±Ë¥•: {e}")
    
    # ÊñπÊ°à2ÔºöÊòæÁ§∫ Mermaid ÊñáÊú¨Ê†ºÂºè
    print("\nüìù ÂõæÁªìÊûÑÔºàMermaid ÊñáÊú¨Ê†ºÂºèÔºâÔºö")
    print("=" * 50)
    mermaid_text = react_graph.get_graph().draw_mermaid()
    print(mermaid_text)
    print("=" * 50)
    
    # ÊñπÊ°à3ÔºöÊòæÁ§∫ÂõæÁöÑËäÇÁÇπÂíåËæπ‰ø°ÊÅØ
    print("\nüîó ÂõæÁªìÊûÑ‰ø°ÊÅØÔºö")
    print("ËäÇÁÇπ:", list(react_graph.get_graph().nodes.keys()))
    print("Ëæπ:", list(react_graph.get_graph().edges))
    
    # ÊñπÊ°à4ÔºöÊèê‰æõÊâãÂä®Ê∏≤ÊüìËØ¥Êòé
    print("\nüí° ÊâãÂä®Ê∏≤ÊüìËØ¥ÊòéÔºö")
    print("1. Â§çÂà∂‰∏äÈù¢ÁöÑ Mermaid ÊñáÊú¨")
    print("2. ËÆøÈóÆ https://mermaid.live/")
    print("3. Á≤òË¥¥ÊñáÊú¨Âà∞ÁºñËæëÂô®‰∏≠Êü•ÁúãÂõæÂΩ¢")
    print("4. ÊàñËÄÖ‰ΩøÁî®ÊîØÊåÅ Mermaid ÁöÑ Markdown ÁºñËæëÂô®")
```
